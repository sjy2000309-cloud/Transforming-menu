```python
def fetch_wikipedia_page(title: str):
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "prop": "extracts",
        "explaintext": 1,
        "titles": title,
        "format": "json",
        "redirects": 1
    }

    headers = {
        "User-Agent": "UoM-TRIM-coursework-benchmark/1.0 (student project)"
    }

    r = requests.get(url, params=params, headers=headers, timeout=20)
    r.raise_for_status()
    data = r.json()

    pages = data["query"]["pages"]
    page = next(iter(pages.values()))

    return {
        "title": title,
        "text": page.get("extract", "")
    }
```

```python
def clean_text(text: str) -> str:
    text = text.replace("\n", " ").strip()
    text = re.sub(r"\s+", " ", text)
    return text

def split_sentences(text: str) -> List[str]:
    text = clean_text(text)
    return [s.strip() for s in re.split(r'(?<=[.!?])\s+', text) if s.strip()]

def strip_title_pronunciation(sentence: str, title: str) -> str:
    s = clean_text(sentence)

    # 去掉类似：
    # "Dim sum (traditional Chinese: ... ) is ..."
    # "Sushi (すし, 寿司...) is ..."
    title_re = re.escape(title)
    s = re.sub(
        rf"^{title_re}\s*\([^)]{{0,180}}\)\s*(?:is|are)\s*",
        f"{title} is ",
        s,
        flags=re.IGNORECASE
    )
    s = re.sub(
        rf"^{title_re}\s*\([^)]{{0,180}}\)\s*",
        f"{title} ",
        s,
        flags=re.IGNORECASE
    )

    return s.strip()

def clean_answer_text(answer: str, title: str) -> str:
    answer = clean_text(answer)
    answer = re.sub(r"^==+\s*[^=]+?\s*==+\s*", "", answer).strip()
    answer = re.sub(r"^=+\s*[^=]+?\s*=+\s*", "", answer).strip()
    answer = re.sub(r"==+\s*[^=]+?\s*==+", " ", answer).strip()
    answer = strip_title_pronunciation(answer, title)
    return answer.strip()
```

```python
def build_sentence_windows(sentences: List[str]) -> List[str]:
    windows = []

    for i, s in enumerate(sentences):
        windows.append(s.strip())

        # 当前句 + 下一句
        if i + 1 < len(sentences):
            windows.append(clean_text(s + " " + sentences[i + 1]))

    # 去重但保持顺序
    deduped = []
    seen = set()
    for w in windows:
        key = w.lower()
        if key not in seen:
            deduped.append(w)
            seen.add(key)
    return deduped

def score_candidate(candidate: str, spec: Dict, title: str) -> int:
    c = clean_answer_text(candidate, title)
    c_lower = c.lower()

    score = 0

    # 先要求基本质量
    if len(c) < 20 or len(c) > 260:
        score -= 3
    if c.count(".") > 1:
        score -= 2

    # 必须和页面主体有一定关系
    if title.lower() in c_lower:
        score += 1

    # required keywords: 每出现一个加分
    for kw in spec["required_keywords"]:
        if kw.lower() in c_lower:
            score += 3

    # forbidden keywords: 每出现一个重罚
    for kw in spec["forbidden_keywords"]:
        if kw.lower() in c_lower:
            score -= 6

    # 一些通用噪声过滤
    noisy_patterns = [
        "etymology",
        "the term comes from",
        "may refer to",
        "pronounced",
        "roman times",
        "working class",
        "this article"
    ]
    for bad in noisy_patterns:
        if bad in c_lower:
            score -= 5

    # 针对不同题型再加一点偏置
    qa_type = spec["qa_type"]

    if qa_type == "origin":
        if any(x in c_lower for x in ["origin", "originated", "introduced", "traced back"]):
            score += 3

    elif qa_type == "used_for":
        if any(x in c_lower for x in ["used", "seasoning", "ingredient", "condiment", "sauce", "spice"]):
            score += 3

    elif qa_type == "served":
        if any(x in c_lower for x in ["served", "carved", "topped", "accompanied", "dipping sauce"]):
            score += 3

    elif qa_type == "when":
        if any(x in c_lower for x in ["breakfast", "brunch", "lunch", "dinner", "summer", "winter", "meal"]):
            score += 3

    return score

def select_best_answer(title: str, text: str, spec: Dict) -> Optional[str]:
    sentences = split_sentences(text)
    windows = build_sentence_windows(sentences)
    search_space = windows[: spec.get("search_limit", len(windows))]

    best_answer = None
    best_score = -10**9

    for cand in search_space:
        cleaned = clean_answer_text(cand, title)
        score = score_candidate(cleaned, spec, title)

        if score > best_score:
            best_score = score
            best_answer = cleaned

    # 最低门槛：避免垃圾题
    if best_answer is None or best_score < 6:
        return None

    return best_answer
```

```python
def build_benchmark(question_specs: List[Dict]) -> List[Dict]:
    benchmark = []
    seen_queries = set()

    # 避免重复下载同一页面
    page_cache = {}

    def add_example(query: str, answer: str):
        if not answer:
            return
        if query in seen_queries:
            return

        benchmark.append({
            "query": query,
            "answer": answer
        })
        seen_queries.add(query)

    for spec in question_specs:
        title = spec["title"]

        try:
            if title not in page_cache:
                page_cache[title] = fetch_wikipedia_page(title)
            doc = page_cache[title]
            text = doc["text"]
        except Exception as e:
            print(f"Skip {title}: {e}")
            continue

        if not text:
            continue

        answer = select_best_answer(title, text, spec)
        if not answer:
            print(f"Skip low-quality item: {title} / {spec['qa_type']}")
            continue

        for template in spec["templates"]:
            add_example(template, answer)

    return benchmark
```

```python
benchmark = build_benchmark(QUESTION_SPECS)

print("Number of examples:", len(benchmark))
print(json.dumps(benchmark, indent=2, ensure_ascii=False))

with open("benchmark.json", "w", encoding="utf-8") as f:
    json.dump(benchmark, f, indent=2, ensure_ascii=False)

print("Saved benchmark.json")
```
