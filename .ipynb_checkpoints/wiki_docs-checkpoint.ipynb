```python
import requests
import re
import json
from typing import List, Dict

# 你可以自己加/删
PAGE_TITLES = [
    "Japanese cuisine",
    "Chinese cuisine",
    "Korean cuisine",
    "Sushi",
    "Kimchi",
    "Ramen",
    "Soy sauce",
    "Tofu",
    "Tempura",
    "Udon",
    "Soba",
    "Peking duck",
    "Hot pot",
    "Dim sum",
    "Bibimbap",
    "Bulgogi",
    "Tteokbokki",
    "Rice",
    "Noodle",
    "Miso",
    "Seaweed",
    "Fish sauce",
    "Ginger",
    "Garlic",
    "Stir frying",
    "Steaming",
    "Deep frying",
    "Fermentation",
    "Braising",
    "Japanese tea ceremony",
    "Korean barbecue",
    "Street food"
]
```

```python
def fetch_wikipedia_page(title: str):
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "prop": "extracts",
        "explaintext": 1,
        "titles": title,
        "format": "json",
        "redirects": 1
    }

    headers = {
        "User-Agent": "UoM-TRIM-coursework-benchmark/1.0 (student project)"
    }

    r = requests.get(url, params=params, headers=headers, timeout=20)
    r.raise_for_status()
    data = r.json()

    pages = data["query"]["pages"]
    page = next(iter(pages.values()))

    return {
        "title": title,
        "text": page.get("extract", "")
    }
```

```python
def clean_text(text: str) -> str:
    text = text.replace("\n", " ").strip()
    text = re.sub(r"\s+", " ", text)
    return text

def split_sentences(text: str) -> List[str]:
    text = clean_text(text)
    return [s.strip() for s in re.split(r'(?<=[.!?])\s+', text) if s.strip()]
```

```python
def get_definition_sentence(title: str, text: str):
    sentences = split_sentences(text[:2500])
    title_lower = title.lower()

    for s in sentences:
        s_clean = clean_text(s)
        if s_clean.lower().startswith(title_lower):
            return s_clean

    for s in sentences:
        s_clean = clean_text(s)
        if title_lower in s_clean.lower():
            return s_clean

    return None

def get_made_from_sentence(title: str, text: str):
    sentences = split_sentences(text)
    patterns = ["made from", "made of", "made with", "consists of", "consisting of"]

    for s in sentences:
        s_clean = clean_text(s)
        s_lower = s_clean.lower()
        if title.lower() in s_lower and any(p in s_lower for p in patterns):
            return s_clean

    return None
```

```python
def build_benchmark(page_titles: List[str]) -> List[Dict]:
    benchmark = []
    seen_queries = set()

    cuisine_keywords = [
        "japanese", "chinese", "korean", "vietnamese", "thai",
        "asian", "east asian", "southeast asian"
    ]

    ingredient_patterns = [
        "made from", "made of", "made with",
        "consists of", "consisting of",
        "contains", "prepared from", "prepared with"
    ]

    preparation_keywords = [
        "served", "prepared", "cooked", "boiled", "fried",
        "steamed", "grilled", "fermented", "seasoned",
        "braised", "stir-fried", "deep-fried"
    ]

    # 更适合保留 ingredients / prepared 问题的页面
    allowed_detailed_pages = {
        "sushi", "kimchi", "ramen", "soy sauce", "tofu", "tempura",
        "udon", "soba", "peking duck", "hot pot", "dim sum",
        "bibimbap", "bulgogi", "tteokbokki", "miso", "fish sauce"
    }

    def add_example(query, answer):
        if not answer:
            return
        answer = clean_text(answer)
        if query not in seen_queries and 20 <= len(answer) <= 220:
            benchmark.append({
                "query": query,
                "answer": answer
            })
            seen_queries.add(query)

    def safe_title(title):
        return title.strip().lower()

    for title in page_titles:
        try:
            doc = fetch_wikipedia_page(title)
            text = doc["text"]
        except Exception as e:
            print(f"Skip {title}: {e}")
            continue

        if not text:
            continue

        sentences = split_sentences(text)
        if not sentences:
            continue

        title_lower = safe_title(title)

        # 1. Definition
        definition = get_definition_sentence(title, text)
        add_example(f"What is {title_lower}?", definition)

        # 2. Made from / made of
        made_from = None
        for s in sentences:
            s_clean = clean_text(s)
            s_lower = s_clean.lower()
            if title_lower in s_lower and any(p in s_lower for p in ingredient_patterns):
                made_from = s_clean
                break
        add_example(f"What is {title_lower} made from?", made_from)

        # 3. Cuisine association
        cuisine_answer = None
        for s in sentences[:15]:
            s_clean = clean_text(s)
            s_lower = s_clean.lower()
            if title_lower in s_lower and any(k in s_lower for k in cuisine_keywords):
                cuisine_answer = s_clean
                break
        add_example(f"What cuisine is {title_lower} associated with?", cuisine_answer)

        # 4. Main ingredients
        if title_lower in allowed_detailed_pages:
            ingredients_answer = None
            for s in sentences:
                s_clean = clean_text(s)
                s_lower = s_clean.lower()
                if title_lower in s_lower and (
                    "ingredient" in s_lower or
                    any(p in s_lower for p in ingredient_patterns)
                ):
                    ingredients_answer = s_clean
                    break
            add_example(f"What are the main ingredients of {title_lower}?", ingredients_answer)

        # 5. Usually prepared
        if title_lower in allowed_detailed_pages:
            prep_answer = None
            for s in sentences:
                s_clean = clean_text(s)
                s_lower = s_clean.lower()
                if title_lower in s_lower and any(k in s_lower for k in preparation_keywords):
                    prep_answer = s_clean
                    break
            add_example(f"How is {title_lower} usually prepared?", prep_answer)

    return benchmark
```

```python
benchmark = build_benchmark(PAGE_TITLES)
bad_queries = {
    "What is japanese cuisine made from?",
    "What is ramen made from?",
    "What are the main ingredients of ramen?",
    "What is hot pot?",
    "What is hot pot made from?",
    "What is bibimbap?",
    "What is bulgogi?",
    "What is bulgogi made from?",
    "What are the main ingredients of bulgogi?",
    "What is rice made from?",
    "What cuisine is rice associated with?",
    "What are the main ingredients of rice?",
    "What is miso made from?",
    "What is seaweed made from?",
    "What are the main ingredients of seaweed?",
    "What cuisine is fish sauce associated with?",
    "What is ginger made from?",
    "What is korean barbecue?",
    "What cuisine is korean barbecue associated with?",
    "What is street food made from?",
    "How is ramen usually prepared?",
    "How is soba usually prepared?",
    "How is dim sum usually prepared?",
    "What is bibimbap made from?",
    "How is bulgogi usually prepared?",
    "How is tteokbokki usually prepared?",
    "What is garlic made from?",
    "What is korean barbecue made from?"
}

benchmark = [
    x for x in benchmark
    if x["query"] not in bad_queries
]

benchmark = [
    x for x in benchmark
    if "lit." not in x["answer"]
    and "== Etymology ==" not in x["answer"]
    and "Muromachi period" not in x["answer"]
    and "precursor of ketchup" not in x["answer"]
    and "served to the king" not in x["answer"]
]
print("Number of examples:", len(benchmark))
print(json.dumps(benchmark[:10], indent=2, ensure_ascii=False))
```

```python
with open("benchmark.json", "w", encoding="utf-8") as f:
    json.dump(benchmark, f, indent=2, ensure_ascii=False)

print("Saved benchmark.json")
```
